```python
from dotenv import load_dotenv
load_dotenv()

import importlib
import sys

# Remove cached modules
if 'task_ingest' in sys.modules:
    del sys.modules['task_ingest']
if 'assignment_rag_helper' in sys.modules:
    del sys.modules['assignment_rag_helper']

from task_ingest import load_data, build_index
from assignment_rag_helper import RAGBase
from openai import OpenAI

documents = load_data()
index = build_index(documents)

openai_client = OpenAI()

assistant = RAGBase(
    index=index,
    llm_client=openai_client,
)

input_tokens, output_tokens, answer = assistant.rag("How does the agentic loop keep calling the model until it stops?")
print(f"Input tokens: {input_tokens}")
print(f"Output tokens: {output_tokens}")
print(f"Answer: {answer}")
```
